```python
import pandas as pd
```

```python
df = pd.read_csv('./data/seoulAccidentInfoList.csv',encoding='utf-8')
```

```python
df
```

```python
address_list = df['시군구'].str.split()
```

```python
address_list.head()
```

```python
df['자치구'] = address_list.str.get(1)
df['행정동'] = address_list.str.get(2)
```

```python
df.head()
```

```python
df = df[['자치구','행정동','사망자수','중상자수','경상자수']]
```

```python
df.groupby(['자치구','행정동']).sum()
```

```python
df.to_csv('서울시_행정동별_교통사고통계.csv',encoding='utf-8')
```
