```python
import pandas as pd
```

```python
house = pd.read_excel('./데이터/가구원수별+가구수_최종.xlsx')
house
```

```python
house.columns =['구','행정동','1인가구']
house
```

```python
index1 = house[house['행정동'] == '소계'].index
print(index1)
```

```python
house = house.drop(index1)
house
```

```python
house = house.reset_index(drop=True)
house
```

```python
popul = pd.read_csv('./데이터/생활인구_202207_동이름_변환.csv')
popul
```

```python
house_popul = pd.merge(house, popul,how='left',on='행정동')
house_popul
```

```python
house_popul.info()
```

```python
house_popul.to_csv('./1인가구_생활인구.csv', encoding='utf-8', index=False)
```
