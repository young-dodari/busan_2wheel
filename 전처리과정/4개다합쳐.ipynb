```python
import pandas as pd
```

```python
df = pd.read_csv('./생활인구1인가구도로율.csv')
df
```

```python
accident = pd.read_csv('./데이터/생활인구_동이름코드_중경상.csv')
accident
```

```python
accident = accident.drop(columns=['행정동코드', '생활인구'])
accident
```

```python
final = pd.merge(df, accident, how='left', on='행정동')
final
```

```python
final.info()
```

```python
final = final.drop(columns=['행정구','동면적','도로율'])
final
```

```python
final.to_csv('./최종이었으면좋겠다.csv', encoding='utf-8', index=False)
```
